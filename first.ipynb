```python
features = ['N', 'P', 'K', 'ph', 'EC', 'S', 'Cu', 'Fe', 'Mn', 'Zn', 'B']

# per crop means
crop_means = dataset.groupby('label')[features].mean().round(2)

# Global means for comparison
global_means = dataset[features].mean().round(2)

# Highlighting significant deviations from global means for each crop
deviations = crop_means - global_means

print("==== Per Crop Mean Values ====")
print(crop_means.to_string())

print("\n==== Deviations from Global Means ====") 
# + means above average, - means below average
print(deviations.to_string())
```
